```python
YR_1 = pd.read_csv("YR_model1.csv")
YR_2 = pd.read_csv("YR_model2.csv")
YR_4 = pd.read_csv("YR_model4.csv")
GLN = pd.read_csv("GLN_2.csv")
GLN_mini = pd.read_csv("GLN_mini_2.csv")
cnn_1 = pd.read_csv("CNN_method1.csv")
cnn_3 = pd.read_csv("CNN_method3.csv")
cnn_1_2= pd.read_csv("method_1_2.csv")
cnn_2 = pd.read_csv("method_2.csv")
cnn_3_2 = pd.read_csv("method_3_2.csv")
```

```python
df_all = pd.DataFrame({"yr_1":YR_1["0"].values, "yr_2":YR_2["0"].values, "yr_4":YR_4["0"].values,
                      "gln": GLN["0"].values, "gln_mini":GLN_mini["0"].values, "cnn_1":cnn_1["0"].values,
                      "cnn_3": cnn_3["0"].values, "cnn_1_2": cnn_1_2["0"].values, "cnn_2":cnn_2["0"].values,
                      "cnn_3_2":cnn_3_2["0"].values})
```

```python



def draw_top2(pcl):
    
    plt.figure(figsize=(5,5), facecolor=[0.9, 0.9, 1])
    le = len(pcl)
    ddf = df_all[pcl] 
    sns.barplot(range(le), np.ones(le), alpha = 0.3)
    sns.barplot(range(le), ddf.iloc[0]/3589.)
    sns.barplot(range(le), ddf.iloc[2]/3589.,bottom=ddf.iloc[0]/3589., alpha = 0.5)
    
    for i in range(le):
        plt.text(range(le)[i]-0.3, (ddf.iloc[0][i] + ddf.iloc[2][i])/3589. + 0.03,"%.4f"%((ddf.iloc[0][i] + ddf.iloc[2][i])/3589.) )
    for i in range(le):
        plt.text(range(le)[i]-0.3, (ddf.iloc[0][i])/3589. + 0.03,"%.4f"%((ddf.iloc[0][i])/3589.) )
    plt.title("Prediction True number for top2",size=15)
    x = range(le)
    plt.xticks(x, ddf.columns, size = 14)
    plt.xlabel("Category", size = 20)
    plt.ylabel("Num", size=20)
    plt.tight_layout()
    plt.show()
```

```python
draw_top2(["cnn_1","cnn_1_2", "cnn_2","cnn_3", "cnn_3_2"])
```
